```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dulieuxettuyendaihoc.csv', index_col='STT')
df.head(10)
```

```python
df.tail(10)
```

Xử lý dự liệu thiếu

```python
df['DT'].isna().sum()
df['DT'].value_counts(dropna=False)
df['DT'].nunique(dropna=False)

```

```python
df['DT'] = df['DT'].fillna(0)
```

Xử lý dữ liệu thiếu – biến T1 (Mean)

```python
df['T1'].isna().sum()
df['T1'].describe()
```

```python
mean_T1 = df['T1'].mean()
df['T1'] = df['T1'].fillna(mean_T1)
```

Xử lý dữ liệu thiếu cho TẤT CẢ biến điểm còn lại

```python
score_cols = [
    'T1','L1','H1','S1','V1','X1','D1','N1',
    'T2','L2','H2','S2','V2','X2','D2','N2',
    'T3','L3','H3','S3','V3','X3','D3','N3'
]

for col in score_cols:
    df[col] = df[col].fillna(df[col].mean())
```

Tạo TBM1, TBM2, TBM3

```python
def calc_TBM(T, L, H, S, V, X, D, N):
    return (T*2 + L + H + S + V*2 + X + D + N) / 10

df['TBM1'] = calc_TBM(df['T1'], df['L1'], df['H1'], df['S1'],
                      df['V1'], df['X1'], df['D1'], df['N1'])

df['TBM2'] = calc_TBM(df['T2'], df['L2'], df['H2'], df['S2'],
                      df['V2'], df['X2'], df['D2'], df['N2'])

df['TBM3'] = calc_TBM(df['T3'], df['L3'], df['H3'], df['S3'],
                      df['V3'], df['X3'], df['D3'], df['N3'])
```

Xếp loại học lực

```python
def xeploai(tbm):
  if tbm >= 8:
    return "Giỏi"
  elif tbm >= 6.5:
    return "Khá"
  elif tbm >= 5:
    return "Trung Bình"
  else:
    return "Yếu"
```

```python
df['XL1'] = df['TBM1'].apply(xeploai)
df['XL2'] = df['TBM2'].apply(xeploai)
df['XL3'] = df['TBM3'].apply(xeploai)

```

```python
def ketqua(row):
    if row['KT'] in ['A', 'A1']:
        return 1 if (row['DH1']*2 + row['DH2'] + row['DH3'])/4 >= 5 else 0
    elif row['KT'] == 'B':
        return 1 if (row['DH1'] + row['DH2']*2 + row['DH3'])/4 >= 5 else 0
    else:
        return 1 if (row['DH1'] + row['DH2'] + row['DH3'])/3 >= 5 else 0
```

```python
import pandas as pd

df = pd.read_csv('/content/dulieuxettuyendaihoc.csv')
```

```python
df['KQXT'] = df.apply(ketqua, axis=1)
df[['KT', 'DH1', 'DH2', 'DH3', 'KQXT']].head()
```
