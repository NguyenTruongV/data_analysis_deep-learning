Thêm header vào dataframe để diễn giải dữ liệu


```python
column_names = ["Id", "Name", "Age", "Weight" ,"m0006","m0612","m1218","f0006","f0612","f1218"]
# đọc file dữ liệu
df = pd.read_csv("patient_heart_rate.csv", header=None, engine="python",on_bad_lines="skip",names = column_names)
# hiển thị một vài dòng dữ liệu đầu tiên ra màn hình
print(df.head())
```

 Xử lý vấn đề một cột lưu hỗn hợp nhiều dữ liệu, ở đây là cột “Name” chứa bao
gồm “Firstname” và “Lastname”, giải pháp là ta sẽ tách ra làm 2 cột  

```python
df[["Firstname","Lastname"]] = df["Name"].str.split(expand=True)
df = df.drop("Name",axis=1)
print(df)
```

Vấn đề 3: Cột Weight có vấn đề về không thống nhất các đơn vị đo lường trong dữ liệu.
Ta sẽ chuyển các đơn vị về thành đơn vị chuẩn “kg”


```python
weight = df[ 'Weight']
for i in range (0 ,len(weight)) :
    x= str(weight [i])
    #Incase ibs is part of observation renove it
    if "lbs" in x[-3:]:
      #Renove the tbs from the value
      x = x[:-3:]
      #Convert string to float
      float_x = float(x)
      #Covert to kgs and store as int y =int (float_x/2.2)
      #Convert back to string
      # đổi lbs -> kgs
      y = round(float_x / 2.2, 2)
      # gán lại
      weight[i] = str(y) + "kgs"
print(df)
```

Vấn đề 4: Vấn đề về xuất hiện dòng dữ liệu rỗng (không có giá trị: NaN). Giải pháp có
thể đưa ra là xóa bỏ


```python
df.dropna(how="all",inplace=True)
print(df)
```

Có nhiều dòng dữ liệu bị trùng lắp thông tin hoàn toàn[fullname, lastname,
age, weight,....], giải pháp đưa ra là chỉ giữ lại một dòng dữ liệu, tuy nhiên giải pháp phải
dựa trên nghiệp vụ của tập dữ liệu và quan sát của người xử lý

```python
df = df.drop_duplicates(subset=["Firstname","Lastname","Age","Weight"])
print(df)
```

 Xuất hiện dữ liệu bị ảnh hưởng bởi lỗi non-ASCII, không định dạng ASCII.
Giải pháp: Tùy vào nghiệp vụ ta có thể: xóa dữ liệu tại đó, thay thế bằng dữ liệu khác
hoặc thay bằng việc đánh dấu bằng một kí tự khác (ví dụ: ‘warning’)

```python
df["Firstname"].replace(
    {r"[^\x00-\x7F]+": "*"},
    regex=True,
    inplace=True
)

df["Lastname"].replace(
    {r"[^\x00-\x7F]+": "*"},
    regex=True,
    inplace=True
)

print(df)
```

Vấn đề 7: “Missing values”, vấn đề này xảy ra tại các cột “Age”, “Weight” và “Heart Rate”. Thiếu dữ liệu (dữ liệu không đầy đủ) là vấn đề xảy ra nhiều trong các nguồn dữ liệu do nhiều nguyên nhân chủ quan lẫn khách quan.Có một vài giải pháp để xử lý vấn đề này, chủ yếu dựa trên kinh nghiệm và nghiệp vụ về tập dữ liệu đó. Một số giải pháp đưa đề xuất từ chuyên gia như sau:
a.	Deletion: Remove records with missing values
b.	Dummy substitution: Replace missing values with a dummy but valid value: e.g.: 0 for numerical values.
c.	Mean substitution: Replace the missing values with the mean.
d.	Frequent substitution: Replace the missing values with the most frequent item.
e.	Improve the data collector: Your business folk will talk to the clients and inform them about why it is worth fixing the problem with the data collector.
……
Yêu cầu:
-	Thống kê thông tin dữ liệu thiếu trên từng biến Age và Weight  
-	Yêu cầu xử lý dữ liệu thiếu như sau: Nếu dòng nào có Age hoặc Weight có dữ liệu thì phần Age hoặc Weight được tính như bên dưới, nếu thiếu cả 2 thông tin thì xóa dòng o Age: Giá trị thay thế là mean của các giá trị trong cột Age


```python
# 1. Thống kê dữ liệu thiếu
print("Missing values before processing:")
print(df[['Age', 'Weight']].isna().sum())

# 2. Xóa dòng thiếu cả Age và Weight
df = df.dropna(subset=['Age', 'Weight'], how='all')

# 3. Tính mean của Age
mean_age = df['Age'].mean()

# 4. Thay thế Age bị thiếu bằng mean
df['Age'].fillna(mean_age, inplace=True)

# 5. (Tuỳ chọn) Nếu muốn xử lý Weight bằng mean thì mở dòng dưới
# mean_weight = df['Weight'].mean()
# df['Weight'].fillna(mean_weight, inplace=True)

# 6. Kiểm tra lại sau xử lý
print("\nMissing values after processing:")
print(df[['Age', 'Weight']].isna().sum())

print("\nData after cleaning:")
print(df)
```

10.	Vấn đề 8: “một cột chứa quá nhiều thông tin cần được phân rã”, như trong bài toán này ta thấy header “m0006” chứa các nội dung bao gồm: m → male, 1218 ~ 12-18 (mm-dd). Còn giá trị thì là kết quả huyết áp.

```python
# Melt the Sex + time range columns into a single column
df = pd.melt(
    df,
    id_vars=['Id', 'Age', 'Weight', 'Firstname', 'Lastname'],
    value_name='Pulserate',
    var_name='sex_and_time'
).sort_values(['Id', 'Age', 'Weight', 'Firstname', 'Lastname'])

# Extract Sex, hours lower and hours upper
tmp_df = df['sex_and_time'].str.extract(r'(\D)(\d+)(\d+)', expand=True)

# Name columns
tmp_df.columns = ['Sex', 'hours_lower', 'hours_upper']

# Create Time column based on hours_lower and hours_upper
tmp_df['Time'] = tmp_df['hours_lower'] + '-' + tmp_df['hours_upper']

# Merge
df = pd.concat([df, tmp_df], axis=1)

# Drop unnecessary columns and rows
df = df.drop(['sex_and_time', 'hours_lower', 'hours_upper'], axis=1)
df = df.dropna()

# Export cleaned data
df.to_csv('outputcleanup.csv', index=False)

print(df)
```
